````python
# replace with file name edit
import os

def replace_text_in_md_files(directory, old_text, text_suffix):
    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.md'):
                file_path = os.path.join(root, file)
                
                # Extract the filename without extension
                filename_without_extension = os.path.splitext(file)[0]
                
                # Construct the new replacement text
                new_replacement_text = filename_without_extension + text_suffix
                
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                
                # Replace the old text with the new text
                updated_content = content.replace(old_text, new_replacement_text)
                
                # Write the updated content back to the file
                with open(file_path, 'w', encoding='utf-8') as f:
                    f.write(updated_content)
                
                print(f"Updated: {file_path}")

# Specify the directory to start the search
directory_to_search = '/path/to/your/directory'

# Specify the old text to replace
old_text_to_replace = '[Diagram]()'

# Define the suffix text to append after the filename
text_suffix = ' text'  # Adjust this as needed

# Run the function
replace_text_in_md_files(directory_to_search, old_text_to_replace, text_suffix)
````

````python
import os

def remove_dot_from_folder_names(directory):
    # Walk through the directory, including subdirectories
    for root, dirs, files in os.walk(directory, topdown=False):
        for dir_name in dirs:
            # Check if the directory name has a dot somewhere in the middle
            if '.' in dir_name and not dir_name.startswith('.') and not dir_name.endswith('.'):
                new_dir_name = dir_name.replace('.', '')
                old_path = os.path.join(root, dir_name)
                new_path = os.path.join(root, new_dir_name)
                
                try:
                    # Rename the directory
                    os.rename(old_path, new_path)
                    print(f"Renamed: {old_path} -> {new_path}")
                except OSError as e:
                    print(f"Error renaming {old_path} to {new_path}: {e}")

# Specify the directory to start the search
directory_to_search = 'c:/users/fpols/documents/github/contemporary physics/demolab/book/book'

# Run the function
remove_dot_from_folder_names(directory_to_search)
````

````python
import os
import re

def find_incorrect_figure_blocks(directory):
    # List to store filenames with incorrect figure blocks
    incorrect_files = []

    # Walk through the directory and its subdirectories
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.md'):
                file_path = os.path.join(root, file)
                
                # Read the file content
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()

                # Find all occurrences of ```{figure} and check for closing ```
                figure_blocks = re.findall(r'```{figure}.*?```', content, re.DOTALL)
                all_figures = re.findall(r'```{figure}', content)

                # Check if the number of {figure} blocks matches the number of closing ```
                if len(figure_blocks) != len(all_figures):
                    incorrect_files.append(file_path)

    return incorrect_files

# Specify the directory to start the search
directory_to_search = 'c:/users/fpols/documents/github/contemporary physics/demolab/book'

# Run the function and print the result
incorrect_files = find_incorrect_figure_blocks(directory_to_search)

# Print the names of files with incorrect figure blocks
if incorrect_files:
    print("Files with incorrect figure blocks:")
    for file in incorrect_files:
        print(file)
else:
    print("All figure blocks are correctly closed.")
````

````python
import os
from docx import Document

def replace_text_in_docx_files(directory, old_text, new_text):
    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.docx'):
                file_path = os.path.join(root, file)
                
                # Load the document
                doc = Document(file_path)
                
                # Iterate over all paragraphs in the document
                for paragraph in doc.paragraphs:
                    if old_text in paragraph.text:
                        paragraph.text = paragraph.text.replace(old_text, new_text)
                
                # Save the updated document
                doc.save(file_path)
                print(f"Updated: {file_path}")

# Specify the directory to start the search
directory_to_search = 'C:/Users/fpols/Documents/SEC'

# Specify the old text to replace and the new text
old_text_to_replace = 'vakcoach'
new_replacement_text = 'werkplekbegeleider'

# Run the function
replace_text_in_docx_files(directory_to_search, old_text_to_replace, new_replacement_text)
````

````python
import os
from docx import Document

def find_files_with_text(directory, search_text):
    files_with_text = []

    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.docx'):
                file_path = os.path.join(root, file)
                
                # Load the document
                doc = Document(file_path)
                
                # Check if the search text is in any paragraph of the document
                for paragraph in doc.paragraphs:
                    if search_text in paragraph.text:
                        files_with_text.append(file_path)
                        print(f"Found '{search_text}' in: {file_path}")
                        break  # Stop checking further once the text is found

    return files_with_text

# Specify the directory to start the search
directory_to_search = 'C:/Users/fpols/Documents/Github/Contemporary Physics/Demolab/book'

# Specify the text to search for
search_text = '()'

# Run the function
found_files = find_files_with_text(directory_to_search, search_text)

# Print the results
print(f"\nTotal files containing '{search_text}': {len(found_files)}")
````

````python
import os

def find_files_with_text(directory, search_text):
    files_with_text = []

    # Walk through the directory
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.md'):
                file_path = os.path.join(root, file)
                
                # Open and read the markdown file
                with open(file_path, 'r', encoding='utf-8') as md_file:
                    content = md_file.read()
                    
                    # Check if the search text is in the file's content
                    if search_text in content:
                        files_with_text.append(file_path)
                        print(f"Found '{search_text}' in: {file_path}")

    return files_with_text

# Specify the directory to start the search
directory_to_search = 'C:/Users/fpols/Documents/Github/Contemporary Physics/Demolab/book'

# Specify the text to search for
search_text = ' http://'

# Run the function
found_files = find_files_with_text(directory_to_search, search_text)

# Print the results
print(f"\nTotal files containing '{search_text}': {len(found_files)}")
````
